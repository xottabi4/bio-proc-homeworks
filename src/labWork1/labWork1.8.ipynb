Necessary imports and configs

```python
from __future__ import division

import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

% matplotlib inline
```

Creating sawtooth signal

```python
f = 5
A = 1
phi = 0
Fs = 1000
ti = 0
tf = 0.5

tff = tf - 1 / Fs
t = np.linspace(ti, tff, round(tff / (1 / Fs)))
sawtoothSignal = np.multiply(A, signal.sawtooth(2 * np.pi * f * t, 1))
# sawtoothSignal=A*sin(2*pi*f*t+phi)

plt.plot(t, sawtoothSignal)
```

Adding gaussian white noise to sawtooth signal

```python
sawtoothSignal_h = sawtoothSignal + A * np.random.randn(1, len(sawtoothSignal))[0]
plt.plot(t,sawtoothSignal_h)
```

Using moving average function to filter signal

```python
windowSize = 10
M = signal.lfilter(np.ones(windowSize) / windowSize, 1, sawtoothSignal_h)
plt.plot(t, M)
```

Creating sin signal

```python
sinSignal = np.multiply(A, np.sin(2 * np.pi * f * t))
plt.plot(t, sinSignal)
```

Adding noise to sin signal

```python
sinSignal_h = sinSignal + A * np.random.randn(1, len(sinSignal))[0]
plt.plot(t,sinSignal_h)
```

Using moving average function to filter signal

```python
windowSize = 10
M = signal.lfilter(np.ones(windowSize) / windowSize, 1, sinSignal_h)
plt.plot(t, M)
```
