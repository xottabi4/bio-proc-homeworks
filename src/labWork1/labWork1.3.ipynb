```python
from __future__ import division

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sc

% matplotlib inline
```

Generate 2 signals: one continuous and second discrete.
Then plot them
After that decimate both and plot again

```python
Fs = 100
ti = 0
tf = 0.5

t = np.arange(ti, tf, 1 / Fs)
n = np.arange(0, len(t))

f = 5
A = 10
phi = 0

# simulation of a sinusoidal signal s(t) at continuous time : s_c
s_c = A * np.sin(2 * np.pi * f * t + phi)

_, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(t, s_c)
# sinusoidal signal sinusoidal s(t) at discrete time : s_d
s_d = A * np.sin(2 * np.pi * f / Fs * n + phi)
ax2.stem(s_d)

# Decimation of a signal
decim = 5  #decimation factor

t_dec = sc.decimate(t, decim)
s_dec = sc.decimate(s_c, decim)

_, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(t_dec, s_dec)
ax2.stem(s_dec)
```
