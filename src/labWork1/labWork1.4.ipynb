```python
from __future__ import division

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sc

% matplotlib inline
```

Generate 2 signals: one continuous and second discrete.
Then plot them
After that decimate both and plot again

```python
Fs = 256
ti = 0
tf = 0.5

t = np.arange(ti, tf, 1 / Fs)
n = np.arange(0, len(t) - 1)

f1 = 100
A1 = 1
phi1 = 0

f2 = 156
A2 = 1
phi2 = 0

f3 = 356
A3 = 1
phi3 = 0

# sinusoidal signal s(t) at continuous time : s_c
s1 = A1 * np.sin(2 * np.pi * f1 * t + phi1)
s2 = A2 * np.sin(2 * np.pi * f2 * t + phi2)
s_c = s1 + s2

# Plot
_, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(t, s1, t, s2)
ax1.set_title('Phase opposite signals s1 and s2, can be retrieved by the relation sin(a-b)')

ax2.plot(t, s_c)
ax2.set_ylim(-1, 1)
ax2.set_title('Sum of signals s1 and s2 => zero sum')
```

```python
# under-sampling sinusoidal signal
s3 = A3 * np.sin(2 * np.pi * f3 * t + phi3)


figure = plt.figure()
ax = plt.plot(t, s1, t, s3)
figure.axes[0].set_title('identical signals s1 and s3, can be retrieved by the relation sin(a+b)')
```
