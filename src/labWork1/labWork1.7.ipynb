```python
from __future__ import division

import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

% matplotlib inline
```

Create triangle pulse. Make it periodical using dirac functions. Have at least 6 replications

```python
triangleWaveLength = 5
t = np.linspace(0, 1, triangleWaveLength)
triangle = signal.sawtooth(2 * np.pi * 5 * t, 0.5)
triangle = triangle + max(triangle)

timeInterval = 0.1
timeLength = round(10 / timeInterval)
totalT = np.linspace(0, 10, timeLength)

triangleWave = np.zeros([1, timeLength])[0]
indicesForWavePlace = list(range(round(timeLength / 2), round(timeLength / 2) + triangleWaveLength))
np.put(triangleWave, indicesForWavePlace, triangle)

plt.plot(totalT, triangleWave)
```

```python
yD = np.zeros([1, len(totalT)])[0]
yD[16] = 1
yD[16 * 2] = 1
yD[16 * 3] = 1
yD[16 * 4] = 1
yD[16 * 5] = 1
yD[16 * 6] = 1
plt.scatter(totalT, yD)
```

```python
newWave = np.convolve(triangleWave, yD)
totalT = np.linspace(0, 10, len(newWave))
plt.plot(totalT, newWave)
```
